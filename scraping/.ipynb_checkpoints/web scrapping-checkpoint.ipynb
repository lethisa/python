```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://www.erajaya.com/company-profile")
c = r.content
```

```python
c
```

```python
soup = BeautifulSoup(c, "html.parser")
```

```python
print(soup.prettify)
```

```python
a = soup.find_all("div",{"class":"container"})
```

```python
a
```

```python
a[0].find_all("a")[0].text
```

```python
for i in a:
    print(i.find_all("a")[0].text)
```

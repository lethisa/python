```python
import pandas
df_geo=pandas.read_json("supermarket.json")
df_geo
```

```python
df_geo["Latitude"]=df_geo["Coordinate"].apply(lambda x: x.latitude if x != None else None)
df_geo["Longitude"]=df_geo["Coordinate"].apply(lambda x: x.longitude if x != None else None)
df_geo
```
